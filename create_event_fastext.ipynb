```python
import pickle
filename = 'event_fasttext_chicago'
outfile = open(filename,'wb')
file = open('eedesc1', 'rb')
data = pickle.load(file)
my_data = {}
file2 = open("out_description1.txt","r")
for keys,values in data.items():
    fasttext = file2.readline()
    print(keys)
    fastext_arr = fasttext.split(' ')
    fasttext_l = []
    for x in fastext_arr[:-1]:
        fasttext_l.append(float(x))
    my_data[keys] = fasttext_l
    print(len(my_data[keys]))
    
file2.close()
pickle.dump(my_data,outfile)
outfile.close()
```
